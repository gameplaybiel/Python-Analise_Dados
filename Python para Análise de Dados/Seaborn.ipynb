**Ciência de Dados para finanças**



O Saborn é uma biblioteca de visualização de dados do Python baseado no Matplotlib

```python
import numpy as np
import pandas as pd
import seaborn as sns

#Definir tema
sns.set_theme(style='darkgrid')
```

```python
#Dados
Base_Dados = sns.load_dataset('tips')

# Verificar
Base_Dados.head()
```

```python
# Renomeando o Base Dados

Base_Dados.rename(columns={
    'total_bill': 'Total Conta',
    'tip': 'Tipo',
    'sex': 'Sexo',
    'smoker': 'Fumante',
    'day': 'Dia Semana',
    'time': 'Período',
    'size': 'Pessoas na Mesa'
}, inplace=True)
```

```python
Base_Dados
```

```python
# Gráfico replot

sns.relplot( x ='Total Conta', y='Gorjeta', data=Base_Dados );
```

```python
sns.histplot(data=Base_Dados, x='Total Conta')
```

```python
sns.histplot(data=Base_Dados, x='Total Conta', hue='Fumante')
```

```python
sns.barplot(data=Base_Dados, x='Sexo', y='Total Conta', hue='Fumante')
```

```python
sns.pairplot(Base_Dados)
```

```python
sns.pairplot(Base_Dados, hue='Sexo')
```

```python
sns.pairplot(Base_Dados, kind='kde')
```

```python
sns.boxplot(data=Base_Dados, x='Dia Semana', y='Total Conta')
```

```python
sns.boxenplot(data=Base_Dados, x='Dia Semana', y='Total Conta', hue='Sexo')
```
