```python
# Bibliotecas
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lendo os dados
BD = pd.read_excel('Dados_Pib.xlsx')

# Verificando
BD.head(10)
```

```python
BD.groupby(by=['Territorialidades', 'Ano']).mean()
```

```python
# Sistema de grids

# Cor fundo de fundo
cor_fundo = '#f5f5f5'

# Incluir cor no fundo
Grid_Graficos = sns.FacetGrid(BD, col='Territorialidades', hue='Territorialidades', col_wrap=4)

# Adicionar gráfico linhas em cada gráfico
Grid_Graficos = Grid_Graficos.map(plt.plot, 'Ano', 'PIB per capita')

# Adicionar uma sombra
Grid_Graficos = Grid_Graficos.map(plt.fill_between, 'Ano', 'PIB per capita', alpha=0.2).set_titles('{col_name} Territorialidades')

Grid_Graficos = Grid_Graficos.set_titles('{col_name}')

# Adicionar um subtitulo
Grid_Graficos = Grid_Graficos.fig.suptitle(
    'Evolução da Renda per capita por Estado. \n Essse relatório foi elaborado no treinado "Python para Análise de Dados".\n',
    fontsize=18
)

# Ajustando
plt.subplots_adjust(top=0.92)
```
