```python
Frase = 'Lorem Ipsum é simplesmente uma simulação de texto da indústria tipográfica e de impressos, e vem sendo utilizado desde o século XVI, quando um impressor desconhecido pegou uma bandeja de tipos e os embaralhou para fazer um livro de modelos de tipos.'
Frase = Frase + 'Ao contrário do que se acredita, Lorem Ipsum não é simplesmente um texto randômico.'
```

```python
Frase
```

```python
# Gerar nuvem de palavras
Nuvem_Palavras = WordCloud(
    background_color = 'black',
    width=1600,
    height=600
).generate(Frase)
```

```python
Nuvem_Palavras
```

```python
# Criar um gráfico
fig, ax = plt.subplots(figsize=(10, 7))
ax.imshow(Nuvem_Palavras, interpolation='bilinear')

ax.set_axis_off()
```
