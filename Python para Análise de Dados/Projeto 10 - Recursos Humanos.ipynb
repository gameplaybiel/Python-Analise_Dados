# **Projeto 10 - Recursos Humanos**

```python
# Libs necessárias

# Libs para a modelagem e matrizes
import pandas as pd
import numpy as np

# Libs para análises gráficas
import matplotlib.pyplot as plt
import seaborn as sns

# Lib para ignorar avisos
import warnings

# Desabilitando os avisos
warnings.filterwarnings('ignore')
```

```python
# Ler os dados
dados = pd.read_csv('Salary_Data.csv')

# Verificar os dados
dados.head()
```

```python
# Renomear os dados
dados.rename( columns={
    'YearsExperience': 'Xp',
    'Salary': 'Renda'
}, inplace=True)
```

```python
# Dimensão
dados.shape
```

```python
# Campos nulos
dados.isnull().sum()
```

```python
# Gráficos nulos
#sns.heatmap(dados.isnull())
sns.heatmap(dados.isnull(), cbar=False)
```

```python
dados.describe()
```

```python
plt.figure(figsize=(10,6))
plt.title('Análise de Renda')
sns.kdeplot(dados['Renda'], color='green')
```

```python
plt.figure(figsize=(10,6))
plt.title('Análise de Xp')
sns.kdeplot(dados['Xp'], color='blue')
```

```python
# Distribuição
plt.figure(figsize=(10,5))
sns.boxplot(dados['Renda'])
```

```python
# Distribuição
plt.figure(figsize=(10,5))
sns.boxplot(dados['Xp'])
```

```python
# Distribuição
plt.figure(figsize=(10,5))
sns.scatterplot(data=dados, x="Renda", y="Xp")
```

```python
plt.figure(figsize=(10,5))
sns.regplot(data=dados, x="Renda", y="Xp")
```

```python
# Correlação
Correlacao = dados.corr('pearson')
plt.figure(figsize=(10,5))
sns.heatmap(Correlacao, annot=True, cbar=True)
```

```python
# Correlação
Correlacao = dados.corr('spearman')
plt.figure(figsize=(10,5))
sns.heatmap(Correlacao, annot=True, cbar=True)
```
