```python
# Bibliotecas
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lendo a base de dados
B_Dados = pd.read_excel('Dados_Empresas_Energia.xlsx')

# Verificando
B_Dados.head()
```

```python
# Séries Temporais - Setar o index
B_Dados.set_index('Data', inplace=True)

# Verificando
B_Dados.head()
```

```python
# Gráfico

# Estilo de gráfico
plt.style.use('seaborn-darkgrid')

# Tamanho
plt.figure(figsize=(16,7))

# Título
plt.title('Análise de ações da empresa de energia', loc='left', fontsize=18, fontweight=0)

# Plot da Petróbras
plt.plot(B_Dados.index, B_Dados['Petrobras'], color='#008c4a', linewidth=4, alpha=0.7)

# Texto da Petróbras
plt.text( B_Dados.index[-1], B_Dados['Petrobras'].tail(1), 'Petrobras', color='#008c4a', size='large', horizontalalignment='left' )

# Plot de todas as colunas
for Coluna in B_Dados.columns[1:]:
  # Plot das outras ações
  plt.plot(B_Dados.index, B_Dados[Coluna], color='gray', linewidth=2, alpha=0.7)
  # Texto da Petróbras
  plt.text( B_Dados.index[-1], B_Dados[Coluna].tail(1), Coluna, color='gray', size='large', horizontalalignment='left' )

# Labels
plt.xlabel('Período')
plt.ylabel('Preço de Fechamento (R$)')
```
