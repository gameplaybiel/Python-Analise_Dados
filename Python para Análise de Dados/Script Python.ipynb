# **29°Estruturas de Classes/Objetos**


```python
class Pessoa:

  #Método construtor
  def __init__(self, nome, idade):
    self.nome = nome
    self.idade = idade

  def Boas_Vindas(self):
    print('Olá sem benvindo: ', self.nome)


  def Recusado(self):
    print('Seu acesso foi recusado!')

    # Função
  def Maior_Idade(self):

    if self.idade >= 18:
       print('O usuário é maior de idade')
       self.Boas_Vindas()
    else:
      print('O usuário é menor de idade')
      self.Recusado()
```

```python
Dados = Pessoa('Odemir', 29)

Dados.Maior_Idade()
```
